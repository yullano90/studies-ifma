1) Faça um algoritmo (em Java) que leia o tempo de duração de um evento em uma fábrica 
expressa em segundos e mostre-o expresso em horas, minutos e segundos.

2) Faça um algoritmo (em Java) que leia as 3 notas de um aluno e calcule a média final deste aluno. 
Considerar que a média é ponderada e que o peso das notas é: 2, 3 e 5, respectivamente



4) Uma fábrica de camisetas produz os tamanhos pequeno, médio e grande, cada uma sendo
vendida respectivamente por 10, 12 e 15 reais. Faça um algoritmo (em Java) em que o usuário 
forneça a quantidade de camisetas pequenas, médias e grandes referentes a uma venda, e a 
máquina informe quanto será o valor arrecadado.


```python
----------------------------------------------------------------------------------------------------

package atv2;
import java.util.Scanner;

public class questoes {
    static public void main(String[] args) {

        
        int tempoEvento;
        System.out.println("\n======== Questão 01 ==========");
        System.out.print("Insira o tempo de duracao do evento em segundos: ");
        Scanner entrada = new Scanner(System.in);
        tempoEvento = entrada.nextInt();
        int segundos = tempoEvento;
        int horas = segundos / 3600;
        segundos %= 3600;
        int minutos = segundos / 60;
        segundos %= 60;
        System.out.print("Tempo total:");
        if (horas > 0) {
            System.out.printf(" %d hora(s)", horas);
        }
        if (minutos > 0) {
            System.out.printf(" %d minuto(s)", minutos);
        }
        if (segundos > 0) {
            System.out.printf(" %d segundo(s)", segundos);
        }

        
        System.out.println("\n\n======== Questão 02 ==========");
        System.out.println("Digite 3 notas do período: ");
        Scanner notaQ1 = new Scanner(System.in);
        float nota01 = notaQ1.nextFloat();
        float nota02 = notaQ1.nextFloat();
        float nota03 = notaQ1.nextFloat();
        float resultado;
        resultado = ((nota01 * 2) + (nota02 * 3) + (nota03 * 5)) / 10;
        System.out.println("\nMedia ponderada = " + resultado);
        
        
        System.out.println("\n\n======== Questão 03 ==========");
        System.out.println("Digite os coeficientes a, b, c, d, e, f: ");
        Scanner inputQ3 = new Scanner(System.in);
        double a = inputQ3.nextDouble();
        double b = inputQ3.nextDouble();
        double c = inputQ3.nextDouble();
        double d = inputQ3.nextDouble();
        double e = inputQ3.nextDouble();
        double f = inputQ3.nextDouble();
        double x = ((c * e) - (b * f)) / ((a * e) - (b * d));
        double y = ((a * f) - (c * d)) / ((a * e) - (b * d));
        System.out.println("\nX = " + x + "\nY = " + y);

        
        System.out.println("\n======== Questão 04 ==========");
        System.out.println("Digite a qtde de camisas G, M e P nessa ordem:");
        Scanner inputQ4 = new Scanner(System.in);
        int g = inputQ4.nextInt();
        int m = inputQ4.nextInt();
        int p = inputQ4.nextInt();
        int totalVenda;
        totalVenda = (g * 10) + (m * 12) + (p * 15);
        System.out.println("\nValor total da vends R$ " + totalVenda);
        
    }
}

----------------------------------------------------------------------------------------------------
```
