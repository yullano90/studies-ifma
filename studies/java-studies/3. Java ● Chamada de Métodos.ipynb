```python
//Declaração de todos os métodos:
----------------------------------------------------------------------------------------------------

package metodos;
import java.util.Scanner;

public class todosOsMetodos {

        //Atividade 03 | Linguagem de Programação II
        //Professor Jose Flavio Gomes Barros
        //Aluno: Iulano Santos 20181SI0006

    public void diasdevida(int idade){
        double dias = idade * 362.25;
        System.out.println("Dias de vida: " + dias);
    }    

   
    public void retangulo(int b, int h){
        int area = b * h;
        int perimetro = b + h;
        System.out.println("Area : " + area);
        System.out.println("Perimetro : " + perimetro);
    }

   
    public int somaantsuc(int num){
        int result = (num-1) + (num+1);
        return result;  
    }
    
    
    public void salario(double horas, double valorhora){
        double salariofinal = horas * valorhora;
        System.out.println("Informe nome do funcionário: ");
        Scanner nome = new Scanner(System.in);
        String funcionario = nome.nextLine();
        String msg = "\nHoras Trabalhadas: " + horas
                + "\nValor da Hora: " + valorhora
                + "\nNome do Trabalhador: " + funcionario
                + "\nSalario Final: " + salariofinal;
        System.out.println(msg);
    }
    
}

----------------------------------------------------------------------------------------------------
```

```python
--------------------------------------------------------------------

package questoes;
import java.text.DecimalFormat;

public class pessoa {

    String nome;
    int idade;
    double peso;
    double altura;
    DecimalFormat df = new DecimalFormat("0.00");

    public void mostrarDados() {
        System.out.println("Nome: " + nome
                + "\nPeso: " + df.format(peso)
                + "\nAltura: " + df.format(altura)
                + "\nIdade: " + idade
        );
    }

    public float calcularIMC() {
        double imc = peso / (altura * altura);
        return (float) imc;
    }
}


--------------------------------------------------------------------
```

```python
--------------------------------------------------------------------

package questoes;
import java.text.DecimalFormat;

public class TestarPessoa {

    public static void main(String[] args) {

        System.out.println("\n==== Dados Pessoa1=====");
        pessoa pessoa1 = new pessoa();
        pessoa1.nome = "Francisco";
        pessoa1.peso = 95.0;
        pessoa1.altura = 1.72;
        pessoa1.idade = 42;
        pessoa1.mostrarDados();

        System.out.println("\n==== Dados Pessoa2=====");
        pessoa pessoa2 = new pessoa();
        pessoa2.nome = "Patrícia";
        pessoa2.peso = 55.0;
        pessoa2.altura = 1.50;
        pessoa2.idade = 23;
        pessoa1.mostrarDados();

        DecimalFormat df = new DecimalFormat("0.00");
        System.out.println("\n==== IMC Pessoa1=====");
        System.out.println("IMC: " + df.format(pessoa1.calcularIMC()));

        System.out.println("\n==== IMC Pessoa2=====");
        System.out.println("IMC: " + df.format(pessoa2.calcularIMC()));
    }
}

--------------------------------------------------------------------
```
