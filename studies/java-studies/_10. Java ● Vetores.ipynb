```python
# Questão 01:
-----------------------------------------------------------------------------
package vetores;
import java.util.Scanner;
public class questaoUm {

    public static void main(String[] args) {
        System.out.println(" ===== Questão 01 ===== ");
        int values[] = new int[6];
        int size = values.length;
        System.out.println("Informe uma sequência de 06 números:");

        for (int i = 0; i < size; i++) {
            Scanner x = new Scanner(System.in);
            int value = x.nextInt();
            values[i] = value;
        }

        for (int j = size - 1; j >= 0; j--) {
            System.out.println(values[j]);
        }
    }
}
```

```python
# Questão 02:
-----------------------------------------------------------------------------
package vetores;
import java.util.Scanner;
public class questaoDois {

    public static void main(String[] args) {
        System.out.println(" ===== Questão 02 ===== ");
        int values[] = new int[10];
        int size = values.length;
        int accumulator = 0;
        System.out.println("Informe uma sequência de 10 números:");
        for (int i = 0; i < size; i++) {
            Scanner x = new Scanner(System.in);
            int value = x.nextInt();
            values[i] = value;
        }
        for (int i = 0; i < size; i++) {
            if (values[i] > 0) {
                accumulator = accumulator + values[i];
            }
        }
        System.out.println(accumulator);
    }
}
```

```python
# Questão 03:
-----------------------------------------------------------------------------
package vetores;
import java.util.Scanner;
public class questaoTres {

    public static void main(String[] args) {
        System.out.println(" ===== Questão 03 ===== ");
        int values[] = new int[10];
        int size = values.length;
        int maxValuePar = 0, minValueImpar = 0;

        System.out.println("Informe uma sequência de 10 números:");
        for (int i = 0; i < size; i++) {
            Scanner x = new Scanner(System.in);
            int value = x.nextInt();
            values[i] = value;
        }

        for (int i = 0; i < size; i++) {
            if (values[i] % 2 == 0) {
                if (values[i] > maxValuePar) {
                    maxValuePar = values[i];
                }
            }
            if (values[i] % 2 != 0) {
                if (values[i] < minValueImpar) {
                    minValueImpar = values[i];
                }
            }
        }

        System.out.println("Maior Valor Par: " + maxValuePar);
        System.out.println("Menor Valor Ímpar: " + minValueImpar);
    }
}
```

```python
# Questão 04:
-----------------------------------------------------------------------------
package vetores;
import java.util.Scanner;
public class questaoQuatro {

    public static void main(String[] args) {
        System.out.println(" ===== Questão 04 ===== ");
        int vet1[] = new int[10];
        int vet2[] = new int[10];
        int vet3[] = new int[10];

        System.out.println("\nVetor 1 - Informe 10 números:");
        for (int i = 0; i < vet1.length; i++) {
            Scanner x = new Scanner(System.in);
            int value1 = x.nextInt();
            vet1[i] = value1;
        }
        System.out.println("\nVetor 2 - Informe 10 números:");
        for (int i = 0; i < vet2.length; i++) {
            Scanner y = new Scanner(System.in);
            int value2 = y.nextInt();
            vet2[i] = value2;
        }
        System.out.println("\nVetor 3 preenchido:");
        for (int i = 0; i < vet3.length; i++) {
            vet3[i] = vet1[i] + vet2[i];
            System.out.println(vet3[i]);
        }
    }
}
```

```python
# Questão 05:
-----------------------------------------------------------------------------
package vetores;
import java.util.Scanner;
public class questaoCinco {

    public static void main(String[] args) {
        System.out.println(" ===== Questão 05 ===== ");
        int values[] = new int[10];
        int size = values.length;
        System.out.println("Informe uma sequência de 10 números:");
        for (int i = 0; i < size; i++) {
            Scanner x = new Scanner(System.in);
            int value = x.nextInt();
            if (value < 0) {
                values[i] = -1;
            } else {
                values[i] = value;
            }
        }

        for (int i = 0; i < size; i++) {
            System.out.println(values[i]);
        }
    }
}
```

```python
# Questão 06:
-----------------------------------------------------------------------------
package vetores;
import java.util.Arrays;
import java.util.Random;

public class questaoSeis {

    public static void main(String[] args) {
        System.out.println(" ===== Questão 06 ===== ");
        int[] vetorAuto = new int[10];
        Random gerador = new Random();
        for (int i = 0; i < vetorAuto.length; i++) {
            vetorAuto[i] = gerador.nextInt(10);
        }
        
        System.out.println(Arrays.toString(vetorAuto));
        int par = 0, impar = 0;
        for (int i = 0; i < vetorAuto.length; i++) {
            if (vetorAuto[i] % 2 == 0) {
                par++;
            } else {
                impar++;
            }
        }
        System.out.println(
                "\n==== Informações do Vetor A ===="
                + "\nNúmeros Pares: " + par
                + "\nNúmeros Ímpares: " + impar
        );
    }
}
```

```python
# Questão 08:
-----------------------------------------------------------------------------
package vetores;
import java.util.Scanner;
public class questaoOito {

    public static void main(String[] args) {
        System.out.println(" ===== Questão 08 ===== ");
        System.out.println("Digite um numero aleatorio: ");
        Scanner entrada = new Scanner(System.in);
        int num = entrada.nextInt();
        int count = 0;

        int vetor[] = new int[]{4, -22, 6, -22, 3, 6, -22, 1, 1, 22};
        for (int i = 0; i < vetor.length; i++) {
            if (num == vetor[i]) {
                count++;
            }
        }
        if (count == 0) {
            System.out.println(
                    num + " não foi localizado no vetor A!"
            );
        } else {
            System.out.println(
                    num + " foi localizado " + count + " X no Vetor A!"
            );
        }
    }
}
```
