```python
-----------------------------------------------------------------------------
# Considere que o Banco tem uma campanha com as regras abaixo adicionadas:
# Clientes que abrirem conta em Julho ganharão R$25,00, em Dezembro R$50,00;
#O código acima deve ser apenas ajustado e as respectivas libs importadas;

package Conta;
import java.time.LocalDateTime;
import java.time.Month;

public class Teste {

    public static double GetSaldoInicial(){
        LocalDateTime now = LocalDateTime.now();
        Month mes = now.getMonth();
        
        if(null == mes)
            return 0;
        else return switch (mes) {
            case DECEMBER -> 50;
            case JULY -> 25;
            default -> 0;
        };
    }
    public static void main(String[] args) {

        try {
            double saldoInicialBanco = GetSaldoInicial();
            double saldoInicialUsuario = 25;
            double saldoInicial = saldoInicialBanco + saldoInicialUsuario;
            Depositar dep = new Depositar(saldoInicial);
            dep.depositar(-2.50);

        } catch (Exception erro) {
            System.out.println(erro);
        }
    }
}

-----------------------------------------------------------------------------
```
