```python
    public static int[] segundosPraTempo(int s) {
        int[] tempoFinal = new int[3];
        int segundos = s;
        int horas = segundos / 3600;
        segundos %= 3600;
        int minutos = segundos / 60;
        segundos %= 60;
        tempoFinal[0] = horas;
        tempoFinal[1] = minutos;
        tempoFinal[2] = segundos;   
        return tempoFinal;
    }

    public static void questaoDez() {
        Scanner entrada = new Scanner(System.in);        
        System.out.println("Insira o numero N de voltas na corrida: ");
        int voltas = entrada.nextInt();  
        int tempos[] = new int[voltas];
        int melhorTemp = 0, melhorVolta = 0, tempoSoma = 0;    
        for(int i = 0; i < voltas; i++) //recebendo os tempos{
            System.out.println("Insira o tempo da volta " + (i+1) + ""
                    + " em segundos: ");   
            tempos[i] = entrada.nextInt();
            //calculando os dados pedidos
            if(i == 0)
                melhorTemp = tempos[i];      
            if(tempos[i] < melhorTemp)
            {
                melhorTemp = tempos[i];
                melhorVolta = i;
            }           
            tempoSoma += tempos[i];
        } 
        
        int media = tempoSoma / voltas;
        int[] tempoFormatado = segundosPraTempo(melhorTemp);
        System.out.print("\nMelhor tempo: ");
        for(int i=0;i<tempoFormatado.length;i++){
            System.out.print(tempoFormatado[i]+":");
        }        
        System.out.println("\nVolta de melhor tempo: " + (melhorVolta+1));
        System.out.println("Tempo médio das " + voltas + " voltas: " + media);
        
```
