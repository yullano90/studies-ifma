```python
//Vamos construir uma Caneta com suas características básicas:
//Modelo, cor, ponta... Tampá-la ou destampá-la com métodos.
--------------------------------------------------------------------

package caneta;
public class bic {

    String modelo;
    String cor;
    float ponta;
    boolean tampada;

    //Com o "this" quem chamar o Status será substituiído por esta função.
    void status() {
        System.out.println("Modelo = " + this.modelo);
        System.out.println("Cor = " + this.cor);
        System.out.println("Ponta = " + this.ponta);
        System.out.println("Tampada = " + this.tampada);
        if (this.tampada == true) {
            System.out.println("Impossível escrever...");
        } else {
            System.out.println("Escrevendo...");
        }
    }

    void tampar() {
        this.tampada = true;
    }

    void destampar() {
        this.tampada = false;
    }
}

--------------------------------------------------------------------
```

```python
//Em seguida, vamos instanciar métodos: 
    
--------------------------------------------------------------------
package caneta;
public class newbics {

    public static void main(String[] args) {
        bic c1 = new bic();
        System.out.println("\n==== Caneta 01 ====");
        c1.modelo = "Bico Fino";
        c1.cor = "Azul";
        c1.ponta = 0.5f;
        c1.tampar();
        c1.status();

        bic c2 = new bic();
        System.out.println("\n==== Caneta 02 ====");
        c2.modelo = "Bico Grosso";
        c2.cor = "Azul";
        c2.ponta = 0.5f;
        c2.destampar();
        c2.status();
    }
}

--------------------------------------------------------------------
```

```python
//Aplicando a visibiliade temos:
--------------------------------------------------------------------
package caneta;

public class bic {

    public String modelo;
    public String cor;
    private float ponta;
    protected int carga;
    private boolean tampada;

    //Com o "this" quem chamar o Status será substituiído por esta função.
    void status() {
        System.out.println("Modelo = " + this.modelo);
        System.out.println("Cor = " + this.cor);
        System.out.println("Ponta = " + this.ponta);
        System.out.println("Tampada = " + this.tampada);
    }

    public void rabiscar() {
        if (this.tampada == true) {
            System.out.println("Impossível escrever...");
        } else {
            System.out.println("Escrevendo...");
        }
    }

    public void tampar() {
        this.tampada = true;
    }

    public void destampar() {
        this.tampada = false;
    }
}

--------------------------------------------------------------------
```

```python
--------------------------------------------------------------------

package caneta;

public class newbics {

    public static void main(String[] args) {
        bic c1 = new bic();
        System.out.println("\n==== Caneta 01 ====");
        c1.modelo = "Bico Fino";
        c1.cor = "Azul";
        //c1.ponta = 0.5f;
        c1.carga = 80;
        c1.tampar();
        c1.status();
        c1.rabiscar();

        bic c2 = new bic();
        System.out.println("\n==== Caneta 02 ====");
        c2.modelo = "Bico Grosso";
        c2.cor = "Azul";
        //c2.ponta = 0.5f;
        c2.carga = 80;
        c2.destampar();
        c2.status();
        c2.rabiscar();
    }
}

--------------------------------------------------------------------
```
