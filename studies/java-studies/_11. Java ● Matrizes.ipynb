```python
# Questão 03:
-----------------------------------------------------------------------------
package matrizes;
import java.util.Scanner;
public class questaoTres {

    public static void main(String[] args) {
        Scanner input = new Scanner(System.in);
        int line_A = 4, column_A = 4;
        int line_B = 4, column_B = 4;
        int M1[][] = new int[line_A][column_A];
        int M2[][] = new int[line_B][column_A];
        int M[][] = new int[4][4];

        System.out.println("\nInforme os elementos de M1:");
        for (int i = 0; i < line_A; i++) {
            for (int j = 0; j < column_A; j++) {
                System.out.print("M1[" + i + "][" + j + "]:");
                M1[i][j] = input.nextInt();
            }
        }
        System.out.println("\nInforme os elementos de M1:");
        for (int i = 0; i < line_B; i++) {
            for (int j = 0; j < column_B; j++) {
                System.out.print("M2[" + i + "][" + j + "]:");
                M2[i][j] = input.nextInt();
            }
        }

        System.out.println("\n=== Matriz M Resultante: ===");
        for (int i = 0; i < line_A; i++) {
            for (int j = 0; j < column_B; j++) {
                for (int k = 0; k < line_B; k++) {
                    M[i][j] += M1[i][k] * M2[k][j];
                }
            }
        }
        for (int i = 0; i < line_B; i++) {
            for (int j = 0; j < column_B; j++) {
                System.out.print(M[i][j] + "  ");
            }
            System.out.print("\n");
        }
    }
}
```

```python
# Questão 04: *
-----------------------------------------------------------------------------
package matrizes;
import java.util.Scanner;
public class questaoQuatro {

    public static void main(String[] args) {
        System.out.println(" ===== Questão 04 ===== ");
        int M[][] = new int[3][3];
        System.out.println("\nPreencha os elementos da Matriz 3X3:");
        Scanner input = new Scanner(System.in);
        //Coletar inputs do usuário:
        for (int i = 0; i < M.length; i++) {
            for (int j = 0; j < M[i].length; j++) {
                System.out.print("Informe matA[" + i + "][" + j + "] = ");
                M[i][j] = input.nextInt();
            }
        }
        //Visualizar cada elemento :
        for (int i = 0; i < M.length; i++) {
            for (int j = 0; j < M[i].length; j++) {
                System.out.println(
                        "Mat A [" + i + "][" + j + "] = " + M[i][j]
                );
            }
        }
        //Contador de elementos repetidos:
        int IndexTest = M[0][0];
        int contador = 0;
        for (int i = 0; i < M.length; i++) {
            for (int j = 1; j < M[i].length; j++) {
                if (M[i][j] == IndexTest) {
                    contador++;
                }
            }
        }
        System.out.println("Elementos repetidos: " + contador);
    }
}
```

```python
# Questão 08: 
-----------------------------------------------------------------------------
package matrizes;
public static void questaOito(){
        int matrizB[][] = {{1,2,3},{4,5,6},{7,8,9}};  
        int maiorNum = 0;        
        List<Integer> mFila = new ArrayList<Integer>();         
        System.out.println("Saída: ");
        
        for (int i = 0; i < matrizB.length; i++) {
            for (int j = 0; j < matrizB.length; j++) {                
                if ((i + (j-1)) == (matrizB.length - 1)){ //abaio da diag principal
                    System.out.print(matrizB[i][j] + ", ");
                    mFila.add(matrizB[i][j]);
                } else if ((i + (j-2)) == (matrizB.length - 1)){
                    System.out.print(matrizB[i][j] + ", ");
                    mFila.add(matrizB[i][j]);
                }                                             
            }
        }
        
        for(int i=0;i < mFila.size();i++)
        {
            if(mFila.get(i) > maiorNum)
                maiorNum = mFila.get(i);
        }        
        System.out.println("\nMaior numero:" + maiorNum);
    }
```

```python
# Questão 12:
-----------------------------------------------------------------------------
package matrizes;
public class questaoDoze {
        int matrizB[][] = {{1,2,3},{4,5,6},{7,8,9}};
        int count = 0, countDois = 0, soma = 0;
        
        for(int i=0;i < matrizB.length;i++){
            for(int j=0;j < matrizB[i].length; j++)
            {                
                if(count == 2)
                {
                    if(countDois < 3)
                    {
                        soma += matrizB[i][j];
                        countDois++;
                    }                    
                    count = 0;
                }
                count++;                
            }            
        }
        System.out.println("Saída: " + soma);        
    }
    
    public static void printMatriz(int[][] vetor, int ln, int col){
        for(int i=0;i < ln;i++){
            for(int j=0;j < col; j++)
            {
                System.out.print(vetor[i][j] + ", ");
            }
            System.out.print("\n");
        }
    }  
}
```
