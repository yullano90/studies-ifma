```python
----------------------------------------------------------------
//Gravação de arquivo e inserção de dados:

#include <stdio.h>
#include <stdlib.h>

int main(void){

    //Declaração do ponteiro para o arquivo + Texto a ser escrito:
    FILE * pointerFile;
    char textFile[20];

    pointerFile = fopen(("gravacao.txt"), "a");

    if(pointerFile == NULL){
        printf("Houve um erro ao abrir o arquivo");
        return 1;
    }
    else{
        printf("Escreva algo no arquivo:\n");
        scanf("%s", &textFile);
        fprintf(pointerFile,"%s\n",textFile);
        fclose(pointerFile);

    printf("Dados gravados com sucesso!\n");
    }

    system("pause");
    return 0;
}
----------------------------------------------------------------
```

```python
----------------------------------------------------------------
//Leitura e Gravação de Arquivos com fputs() e fgets();

#include <stdio.h>
#include <stdlib.h>
#include <string.h>
#include <locale.h>
//#include <windows.h>

int main(void) {
/*    //Regionalization Command
    setlocale(LC_ALL, "Portuguese");
    //Program Title
    SetConsoleTitle("Manipulando Arquivos e pastas");

    if(CreateDirectory("DirectoryTestFiles", NULL)){
        printf("Arquivo criado com sucesso!\n\n");
    }
    else{
        printf("Erro ao criar pastas e arquivos!\n\n");
    }
*/
  
  FILE *pointerFile;
  int r;
  char texto_str[51];

  // file opening
  pointerFile = fopen("gravacao-fputs.txt", "a");

  //Disk Open or Write Error Check
  if (pointerFile == NULL) {
    printf("houve um erro ao tentar abrir o arquivo!");
    exit(1);
  }

  //Recording with fputs()
  r = fputs("revisao de manipulacao de arquivos 4\n", pointerFile);

  //Text write error checking
  if (r == EOF) {
    printf("Houve um erro ao gravar dados no arquivo!");
  } else {
    printf("Dados gravados com sucesso!\n");
  }
  fclose(pointerFile);

  //Reading with fgets()
  printf("\n=== Leitura de dados gravados no arquivo ===\n");
  pointerFile = fopen("gravacao-fputs.txt", "r");
  while (fgets(texto_str, 51, pointerFile) != NULL) {
    printf("%s\n", texto_str);
  }
  fclose(pointerFile);

  system("pause");
  return 0;
}
----------------------------------------------------------------
```
