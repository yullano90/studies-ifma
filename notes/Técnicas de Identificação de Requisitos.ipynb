## Introdução à abordagens de identificação de requisitos:

O levantamento de requisitos é a atividade que starta o desenvolvimento de um Software e também está em todas as etapas da Engenharia de Requisitos.

Vide abaixo as atividades, segundo Sommerville (2003), que estão no processo genérico de levantamento e análise:
- Compreensão do Domínio da Aplicação;
- Coleta de Requisitos junto aos Stakholders;
- Classificação de Grupos de Requisitos;
- Resolução de Conflitos que vem à tona;
- Definição das Requisitos Prioritários;
- Verificação de Consistência e Concordância de Requisitos;


## Dificuldades encontradas:

Segundo Pompilho (1995), entre as razões que contribuem para o baixo grau de satisfação de usuários para os sistemas, podemos destacar:
- Falta de Técnica adequada para extrair requisitos do sistema;
- Falha na clareza da descrição de requisitos do sistema;
- O usuário principal do sistema não sabe o que quer;



## Técnicas de Levantamento de Requisitos:

● Levantamento Orientado a Pontos de Vista => Diferentes pontos de vista são discutidos e um framework é ofercido;

● Etnografia => Técnica de observação organizacional e cultural, promovendo familiarização da história do sistema; 

●  Workshops => Essa técnica tem por objetivo acionar o trabalho em equipe. Há um facilitador neutro que conduz e promove a discussão entre os vários mediadores.

●  Prototipagem => Tem por objetivo explorar aspectos críticos dos requisitos de um produto, implementando de forma rápida um pequeno subconjunto de funcionalidades deste produto.

●  Entrevistas => Técnica tradicional mais simples que produz bons resultados na fase inicial de obtenção de dados. Convém que o entrevistador dê margem ao entrevistado para expor as suas idéias.

●  Questionários => Indicado nos casos em que os grupos de usuários diversos estejam distribuídos em locais diferentes e tenham contribuição relevante na construção do software, deve ser acompanhado por uma carta explicativa com a relevância da pesquisa.

● Brainstorming => Tempestade de ideias, objetivando uma forte ideação e depois as mais relevantes são filtradas.

● JAD => (Joint Application Design) é uma técnica para promover cooperação, entendimento e trabalho em grupo entre os usuários desenvolvedores.
O JAD facilita a criação de uma visão compartilhada do que o produto de software deve ser.
