## Comunidades de Software e sua importância:

A comunidade de software livre é composta por membros altamente colaborativos que, mesmo estando envolvidos em um mesmo projeto, não se conhecem pessoalmente já que na maioria dos casos se encontram separados por grandes distâncias geográficas, interagindo apenas através de ferramentas disponíveis na Internet. Isso é verdade para a maioria dos casos, mas cabe lembrar que existem exceções, não é incomum que membros de um projeto deste tipo façam todos parte de uma única empresa, que pode vir inclusive a começar um projeto como um software comercial e mais tarde transformá-lo em um software livre.

Podemos citar como exemplo deste último caso, sistemas famosos como o Mozilla, o compilador Java Jikes da IBM e o Java Development Kit da Sun. 

Os retornos obtidos por membros destes projetos incluem o reconhecimento dentro da comunidade de software, a melhoria do 
software em questão (já que usualmente o membro é usuário deste software) e o fato de estar contribuindo para
a expansão dos softwares livres no mercado atual.

## Comunidade CISL 

O Comitê Técnico de Implementação de Software Livre do Governo Federal é incentivador do projeto DEMOSEILLE, software escolhido para abordagem nesse trabalho da disciplina Software Livre. O Demoiselle foi desenvolvido internamente em 2008 no SERPRO (serviço federal de processamento de dados e patrocinador da comunidade Demoiselle), e posteriormente disponibilizado pra comunidade.




## Framework Demoiselle: 

- Solução em comunidade de software livre para desenvolvimento ágil e flexível com linguagem JAVA;
- O software veio crescendo na mesma proporção que o JEE (Java Enterprise Edition);
- Apoia a desenvolvimentos de sistemas em especial na linguagem Java (com artefatos também para outras linguagens);
- O SERPRO usa o Demoiselle e a comunidade avalia melhorias constantemente.


## Ecossistema Demoiselle

Conjunto de Soluções com duas grandes divisões:

Process:
- Desenvolvimento:
- Negócio:

Projetos:
- Framework;
- Componente;
- Infra;
- Tools;
    
## Comunidade DEMOISELLE:

Objetivos: 
- Fomentar o uso de soluções;
- Evolução e desenvolvimento mais rápido do processo;
- Apoio ao usuário e colaborador;
- Garantir a continuidade do projetos;
- Estabelecer critérios de funcionamento;
- Independência Tecnológica;

A comunidade tem contribuído:
- Testando o código;
- Relatando e corrigindo bugs;
- Submetendo componentes e melhorias;

Segue as 04 liberdades estabelecidas pela FSF - Free Software Foundation:
- Liberdade pra executar o programa(para qualquer propósito);
- Liberdade para estudar o software;
- Liberdade para redistribuir cópias do programa de modo que você possa ajudar ao seu próximo;
- Liberdade de modificar o software
