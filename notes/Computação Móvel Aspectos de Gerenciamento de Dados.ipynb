## 04 Atividade - Computação Móvel Aspectos de Gerenciamento de Dados 

01) O banco de dados moveis pode ser definido como um “armário” eletrônico que coleciona arquivos de dados computadorizados, facilitando o acesso do usuário a qualquer informação contida nele, com poucas ou nenhuma restrição de tempo e lugar. A sua arquitetura se divide em duas, de computadores moveis e computadores fixos. Os computadores moveis: 
- São interligados em rede através de ligações sem fio, que permitem a mobilidade estabelecendo uma conexão virtual de qualquer localização dentro de uma área geográfica;
- Os computadores fixos: As estações e conectividades não mudam, essas estações realizam a comunicação entre a rede fixa e computadores móveis.

---------------------------------------------------------------------------------------------------------

02)  - SyBase Ultralite, DB2 Everyplace, SQLite

---------------------------------------------------------------------------------------------------------

03) É um movimento entre duas estações adjacentes que garante a transação de dados enquanto o usufrutuário se desloca entre uma célula e outra, elas mantem a conectividade da unidade móvel. É um processo realizado de forma transparente ao usuário.

---------------------------------------------------------------------------------------------------------

04) O processo de detecção deve ser feito bem antes da pera de conexão, o tempo necessário para o processo pode causar atrasos na entrega de pacotes e afeta o desempenho do TCP e aplicações em tempo real, pode haver do processo não se completar por falta de recursos na BS.

---------------------------------------------------------------------------------------------------------

05) Pode ocorrer quando uma unidade móvel se encontra fora da área de cobertura das células. A desconexão pode ser classificada como voluntaria e involuntária. A voluntária não se caracteriza por uma falha pois acontece devido a mobilidade e o poder de bateria dos aparelhos. A involuntária ou forçada acontece quando a unidade móvel se encontra em uma área fora de cobertura.

---------------------------------------------------------------------------------------------------------

06) No contexto do gerenciamento de dados moveis, a difusão de dados, consiste no envio de mensagens enviada através de uma estação de suporte móvel, para todas as unidades moveis que se encontram em uma determinada célula.

---------------------------------------------------------------------------------------------------------

07) Largura de banda alta, mobilidade, consumo de energia, tempo de transmissão,
execução da consulta, etc.

08) Pois pela capacidade do ambiente móvel, proporcionar a mobilidade, ele oferece uma recuperação de dados a qualquer hora e sem limites geográficos. Esse processo é responsável pela consistência do banco de dados após alguma falha no sistema.

---------------------------------------------------------------------------------------------------------
09) As técnicas consistem em gravar um período de “restauração”, caso venha dar algum problema no banco e o mesmo tenha que ser ecuperado.
