### Importância da Linha Visada entre antenas:

A linha visada é fundamental para balizar como uma antena consegue enxergar perfeitamente à outra (quando não há obstáculos), assim não ocorrerá problema na comunicação entre transmissor e receptor, no circuito RF, fazendo com que o link não seja afetado.

Por outro lado, quando existe este obstáculo, é necessário evitar o grau de interferência, o que é chamado de Zona de Fresnel. Da mesma forma, pode ser definida como uma série de elipses concêntricas em torno da linha visada, ou seja, a Zona de Fresnel determina a área em torno da linha de visada na qual poderá introduzir interferência no sinal, caso ele seja bloqueado. Se desobstruídas, as ondas de rádio viajam em linha reta a partir do transmissor para o receptor.

No site que foi feita a simulação (no vídeo) temos alguns parâmetros indispensáveis para dimensionar e observar:
- Modelos de equipamento;
- Altura da torre;
- Antena (Se for integrada não é possível alterar);
- Ganho
- Perda (Nível de ruído);
- Localização (Coordenadas);
- Tipos de Temperatura (Clima Tropical, Frio etc);

Assim torna-se possível a detecção de obstrução de links sem agredir sinal, modulação e taxa de transferência.
